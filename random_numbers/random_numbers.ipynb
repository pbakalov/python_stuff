```python
from utils import  power_law_cdf, power_law_pdf, power_law_rngen
    
    
# rns = []

x0 = 0.01
alpha = 1.05

xmax = 3.
nbins = 400
manual_hist = zeros(nbins, float)
bin_size = (xmax-x0)/nbins

for i in xrange(1000000):
    x = power_law_rngen(x0,alpha)
#     rns.append(x)
    if x<xmax:
        manual_hist[int((x-x0)/bin_size)] += 1
    
    
# plt.hist(rns, range = (x0,xmax),bins = nbins, density=True,log = True , stacked = True)
# print max(rns)
# print min(rns)
```
